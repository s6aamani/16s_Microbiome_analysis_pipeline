```r
library("genefilter")
library(phyloseq)
library("qiime2R")
library(PMA)
library(ade4)
library(ggplot2)
library(ggrepel)
```

```r
setwd('/Users/alchemist/Aakash/BonnGut/Microbiome_pipeline')
workdir = getwd()
```

```r
#
pseq<-qza_to_phyloseq(
    features=paste(workdir,"/feature_table/table.qza",sep=""),
    tree=paste(workdir,"/phylo_tree/rooted_tree.qza",sep=""),
         paste(workdir,"/taxonomy/tax_ezbiocloud.qza",sep=""),
    paste(workdir,"/metadata/metadata.tsv",sep="")
    )
meta_data = data.frame(sample_data(pseq))
metab = meta_data[,c("XXX","YYY","ZZZ")]
head(meta_data)
```

```r
head(metab)
```

# Genus

## Agglomerate taxa

```r
#
#
microbe = tax_glom(pseq, taxrank="Genus") 
microbe <- core(microbe, detection = 0.0001, prevalence = .5) #default filtering
microbe <- prune_taxa(taxa_sums(microbe) > 4, microbe)
microbe <- filter_taxa(microbe, filterfun(kOverA(3, 2)), TRUE)
dataMatrix <- data.frame(otu_table(microbe))
rownames(dataMatrix) <- as.data.frame(tax_table(microbe))$Genus
dataMatrix
```

## Some Preprocessing

```r
keep_ix <- rowSums(metab == 0) <= 3
metab <- metab[keep_ix, ]
# metab <- log(metab, base = 10)
X <- dataMatrix #otu_table(microbe)
# X[X > 50] <- 50
dim(X)
metab = t(metab)
```

```r
cca_res <- CCA(t(X),  t(metab), penaltyx = .15, penaltyz = .15)
```

```r
cca_res
```

```r
combined <- cbind(t(X[cca_res$u != 0, ]),
                 metab[cca_res$v != 0, ])
pca_res <- dudi.pca(combined, scannf = F, nf = 3)
```

```r
combined
```
